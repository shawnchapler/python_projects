#Problem 1(a).  Unionized workers.

These problems use the 1985 Current Population Survey data, ```cps.csv```, from last week.

Create a pandas data frame to hold the CPS data.  Write a function that computes the average hourly wage of a unionized worker by using control flow to iterate through each row of the data frame.

````python
import pandas as pd
import numpy as np
````

````python
#load File 
cps = pd.read_csv('C:/Users/smc/Documents/GitHub/ds710summer2017assignment9/cps.csv', skipinitialspace=True)

#Review Data
cps.describe()
cps
````

````python
#Identify Unique values for Union
cps['union'].unique()
````

````python
#Create function to submset union employees and calculate the average hourly wage on the data frame
def avg_hourly_wage(x):
    """ return avg hourly wage for union rows"""
    total_wages=0
    count=0
    index=0
    for i in x['union']:
        if i == "Union":
            total_wages=total_wages+x.wage[index]
            count=count+1
            avg_hrly=round(total_wages/count, 2)
        index=index+1
    return avg_hrly 

````

````python
#Confirm output
print(avg_hourly_wage(cps))
````

#Problem 1(b).  More unionized workers.

Write a function that takes a row of a data frame as input.  Your function should return 0 if a worker is not unionized, and the worker's hourly wage if the worker is unionized.  Apply this function to your data frame, then use the resulting series to compute the average hourly wage of a unionized worker.

````python
# We define a function to test whether a row is union 

def is_union(row):
    """ return wage if row is Union else return 0"""
    
    if row['union'] == "Union":
        return row['wage']
    else:
        return 0
````

````python
#apply function to cps dataset along rows
cps_union=cps.apply(is_union, axis=1)
````

````python
#Review output
cps_union
````

````python
#compute average hourly wage for union employees using dataframe from cps_union function
count_ul = 0
sum_wage_ul=0
avg_hrly_ul=0
for i in cps_union:
    if i != 0.00:
        count_ul = count_ul + 1
        sum_wage_ul=sum_wage_ul + i
        avg_hrly_ul=sum_wage_ul/count_ul
    avg_hourly_wage_ul = round(avg_hrly_ul, 2)

````

````python
#Confirm output
print(avg_hourly_wage_ul)
````
