# Problem 1(a).  College data.

The data set usnews.csv contains data from the 1995 US News and World Report on colleges and universities in the US.  

(Dataset:  “US News College data,” submitted by Robin Lock, St. Lawrence University.   Dataset obtained from the Journal of Statistics Education (http://www.amstat.org/publications/jse). Accessed 23 June 2015. Used by permission of author.)

Metadata:  http://www.amstat.org/publications/jse/datasets/usnews.txt


The data set contains asterisks which indicate missing data.  Use pandas to create a data frame which replaces the asterisks by the ```NaN``` symbol. Also look out for extra spaces.  

````python
import pandas as pd
import numpy as np
````

````python
#load File 
us_colleges = pd.read_csv('usnews.csv', skipinitialspace=True)

#Review Data
us_colleges
us_colleges.describe()

#Create data frame and replace asterisks and do something with the extra spaces.  Iterate over all columns for *
def remove_asteriks(v):
    """ return np.nan if the value is an asterisk"""
    if v == "*":
        return np.nan
    else:
        return v

#apply to dataset
us_colleges = us_colleges.applymap(remove_asteriks)

#review dataset
us_colleges
    
````

#Problem 1(b).  Categorizing universities.

Create a new column in your data frame which describes whether a university is public or private, using the words "Public" and "Private" instead of numbers.  (It may be helpful to consult the metadata file.)


````python
#Identify Unique values for Public/private
us_colleges['Public/private'].unique()

#Create new column in data frame which describes whether a university is public or private
PubPriv_dict = {1:'Public', 2:'Private'}

#Create a lambda function which calls the PubPriv_Dict and returns a series object
us_colleges['Public/private'].map(lambda s: PubPriv_dict.get(s))
            
#Create new column and map PubPriv_Dict series to it
us_colleges['Public/private Desc'] = us_colleges['Public/private'].map(lambda s: PubPriv_dict.get(s))
us_colleges
````

#Problem 1(c).  

The interquartile range is the 3rd quartile minus the 1st quartile.  Create columns recording the interquartile range for math and verbal SAT scores.

````python
list(us_colleges)
````

````python
#Handle NaN's in field prior to computation
#us_colleges['Third quartile - Math SAT'].fillna(0)
#us_colleges['First quartile - Math SAT'].fillna(0)
#us_colleges['Third quartile - Verbal SAT'].fillna(0)
#us_colleges['First quartile - Verbal SAT'].fillna(0)
us_colleges['Third quartile - Math SAT']
````

````python
#Create function to compute interquartile range for Math SAT to apply to each row in the data frame
def compute_interquartileMath(row):
    a=row['Third quartile - Math SAT']
    b=row['First quartile - Math SAT']
    c=float(a)-float(b)
    return c

us_colleges.apply(compute_interquartileMath, axis=1)
#Create function to compute interquartile range for Verbal SAT to apply to each row in the data frame
def compute_interquartileVerbal(row):
    a=row['Third quartile - Verbal SAT']
    b=row['First quartile - Verbal SAT']
    c=float(a)-float(b)
    return c



# Create new columns to hold Interquartile values
us_colleges['Interquartile - Math SAT'] = us_colleges.apply(compute_interquartileMath, axis=1)
us_colleges['Interquartile - Verbal SAT']= us_colleges.apply(compute_interquartileVerbal, axis=1)

us_colleges
````
